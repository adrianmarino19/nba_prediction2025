```python
# --- 2.1  start from the wide 'game' table you loaded ---------------
games_w = pd.read_sql("SELECT * FROM game", conn)

# --- 2.2  create a helper to pick HOME vs AWAY columns -------------
home_cols = [c for c in games_w.columns if c.endswith("_home")]
away_cols = [c for c in games_w.columns if c.endswith("_away")]

def melt_side(df, side, cols):
    out = df[["game_id", "game_date"] + cols].copy()
    rename = {c: c.replace(f"_{side}", "") for c in cols}
    out = out.rename(columns=rename)
    out["is_home"] = 1 if side == "home" else 0
    return out

games_home = melt_side(games_w, "home", home_cols)
games_away = melt_side(games_w, "away", away_cols)

games = pd.concat([games_home, games_away], ignore_index=True)
```

```python
# Fixing team names and removing All-Star game entries.

CODE_MAP = {
    # Mismatches you saw
    "PHX": "PHO",   # Phoenix
    "BKN": "BRK",   # Brooklyn Nets
    "GOS": "GSW",   # Golden State
    "SAN": "SAS",   # San Antonio
    "UTH": "UTA",   # Utah
    "PHL": "PHI",   # 1940s Sixers label in Kaggle
    # Already-known legacy moves
    "NOH": "NOP", "NOK": "NOP",
    "CHA": "CHO", "NJN": "BRK", "SEA": "OKC"
}

games["team_abbreviation"] = games["team_abbreviation"].replace(CODE_MAP)

# games = games.query("season_type != 'All Star'")
```

```python
# ELO

elo = pd.read_csv("data/nbaallelo.csv", parse_dates=["date_game"])

elo = elo.rename(columns={
    "team_id": "team_abbreviation",
    "date_game": "game_date",
    "elo_i": "elo_pre",
    "elo_n": "elo_post"
})
```

```python
# how many NAs do we have?

games["game_date"] = pd.to_datetime(games["game_date"], errors="coerce")

df = (games
      .merge(elo[["team_abbreviation","game_date","elo_pre"]],
             on=["team_abbreviation","game_date"],
             how="left"))

print("NA share after code patch:", df["elo_pre"].isna().mean())
```
